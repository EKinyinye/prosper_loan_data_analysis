## Investigation Overview




The goal was to check of the Loan Status variable and invetigate it's relationship with other variables such as the loan original amount, borrower's rate of interest, employment status, stated monthly income, mothly loan payment among others. The key insights as explained in the below investigation will be on Term, loan original amount, borrower's rate and employment status.


## Dataset Overview

> This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df= pd.read_csv('loan.csv')
````

````python
df.info()
````

## (Visualization 1)


A bar plot showing the distribution of the LoanStatus variable. Most of the loans are currently stilled owned, followed by completed loan. Few also fall in the other categories with some close to zero in frequency.



````python
base_color = sb.color_palette()[0]
sb.countplot(data=df, x='LoanStatus', color = base_color)
plt.xticks(rotation=90)
plt.title('Distribution of the LoanStatus');
````

## (Visualization 2)



A box plot showing the relation between two variables, that is, Employment Status and Loan Original Amount. Most of the borrowers fall under the category of Employed followed by Self-Employed. Most of these could be people in need of income booster and probably pay the loan whenever they receive their payments.

````python
sb.boxplot(data=df, y='EmploymentStatus', x='LoanOriginalAmount',color=base_color)
plt.title('Relationship between EmploymentStatus and MonthlyLoanPayment');
````

## (Visualization 3)
A scatter plot showing the relationship between the three variables, LoanOriginalAmount, MonthlyLoanPayment and Term of Payment. The monthly repayment increases as the original loan amount increases. It's also quite clear that most of the loans are repayable by 36 months followed by 60 and lastly 12.

````python
g = sb.FacetGrid(data = df, hue = 'Term', size = 7.5,
                 palette = 'colorblind')
g.map(plt.scatter, 'LoanOriginalAmount', 'MonthlyLoanPayment')
g.add_legend()
plt.title('Relationship between LoanOriginalAmount, MonthlyLoanPayment, and Term');
````

## (Visualization 4)

The scatter plot above shows the relationship between MonthlyLoanPayment, BorrowerRate and LoanOriginalAmount variables. Monthly payments increases with the loan original amount. This is als the case with the borrower's rate. Although some of the high amounts loans appear to have high amounts in terms of monthly payments which in terms leads to lower rates of intersest.

````python
plt.scatter(data = df, y = 'MonthlyLoanPayment', c = 'BorrowerRate', x = 'LoanOriginalAmount',
            cmap = 'mako_r')
plt.colorbar()
plt.xlabel('LoanOriginalAmount')
plt.ylabel('MonthlyLoanPayment')
plt.title('Relationship between MonthlyLoanPayment BorrowerRate and LoanOriginalAmount');
````

From the above investigations carried out, most of the loans are owned by employed people followed by self-employed. This could be as a result of in need of income boosters which they would in turn pay with their monthly payments from the employers or from the business profits respectively. The monhly payment depends on the original loa amount borroed and the rate of interest. 

Loans with lower amount in terms of monthly payment tends to take longer in terms of payment period and vice versa.
